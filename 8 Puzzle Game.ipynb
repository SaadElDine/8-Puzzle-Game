### Determine whether the initial puzzle state is solvable or not!

```python
# Count inversions to determine whether the initial puzzle state is solvable or not
def count_inversions(puzzle):
    flat_puzzle = [tile for tile in puzzle if tile != 0]
    inversions = 0

    for i in range(len(flat_puzzle)):
        for j in range(i + 1, len(flat_puzzle)):
            if flat_puzzle[i] > flat_puzzle[j]:
                inversions += 1

    return inversions

# Function to check if a puzzle is solvable
def is_solvable(puzzle):
    inversions = count_inversions(puzzle)
    return inversions % 2 == 0
```

# 1.BFS Algorithm

```python
from collections import deque
import time

# Breadth-First Search (BFS) algorithm to solve the 8-puzzle
def bfs(initial_state):
    goal_state = [0, 1, 2, 3, 4, 5, 6, 7, 8]

    frontier = deque()  # Create a deque to store the states to be explored
    frontier_states = set()  # Create a set to keep track of states in the frontier
    frontier.append((initial_state, []))  # Initialize the frontier with the initial state and an empty path
    frontier_states.add(tuple(initial_state))  # Add the initial state to the set of frontier states

    explored = set()  # Create a set to keep track of explored states
    start_time = time.time()  # Record the start time for measuring execution time

    search_depth = 0  # Initialize the search depth

    while frontier:  # Continue searching as long as there are states in the frontier
        current_state, path = frontier.popleft()  # Get the state and path from the front of the frontier

        if current_state == goal_state:  # Check if the current state is the goal state
            end_time = time.time()  # Record the end time
            execution_time = end_time - start_time  # Calculate the running time
            return [initial_state] + path, len(path), len(explored), search_depth, execution_time  # Return path, cost, nodes expanded, search depth, and running time

        explored.add(tuple(current_state))  # Add the current state to the set of explored states

        empty_tile_index = current_state.index(0)  # Find the index of the empty tile (0)
        x, y = empty_tile_index // 3, empty_tile_index % 3  # Calculate its coordinates in the 3x3 grid

        # Define possible moves: Up, Down, Left, Right
        moves = [(x - 1, y), (x + 1, y), (x, y - 1), (x, y + 1)]

        for move_x, move_y in moves:
            if 0 <= move_x < 3 and 0 <= move_y < 3:  # Check if the move is within the grid
                new_state = current_state[:]  # Create a copy of the current state
                new_empty_tile_index = move_x * 3 + move_y  # Calculate the index of the new empty tile
                new_state[empty_tile_index], new_state[new_empty_tile_index] = new_state[new_empty_tile_index], new_state[empty_tile_index]  # Swap the empty tile with a neighboring tile

                new_state_tuple = tuple(new_state)  # Convert the new state to a tuple

                # Check if the new state has not been explored and is not in the frontier
                if new_state_tuple not in explored and new_state_tuple not in frontier_states:
                    frontier.append((new_state, path + [new_state]))  # Add the new state to the frontier with the updated path
                    frontier_states.add(new_state_tuple)  # Add the new state to the set of frontier states
                    search_depth = max(search_depth, len(path))  # Update the search depth
        # Check if the elapsed time exceeds 5 minutes
        if time.time() - start_time > 300:  # 300 seconds = 5 minutes
            return None  # Return None if the execution time exceeds 5 minutes
```

### 2. DFS Algorithm

```python
def dfs(initial_state):
    goal_state = [0, 1, 2, 3, 4, 5, 6, 7, 8

    stack = [(initial_state, [])]  # Create a stack with the initial state and an empty path
    explored = set()  # Create a set to keep track of explored states
    stack_states = set()  # Create a set to keep track of states in the stack
    start_time = time.time()  # Record the start time for measuring execution time
    search_depth = 0  # Initialize the search depth

    while stack:  # Continue searching as long as the stack is not empty
        current_state, path = stack.pop()  # Pop the state and path from the top of the stack

        if current_state == goal_state:  # Check if the current state is the goal state
            end_time = time.time()  # Record the end time
            execution_time = end_time - start_time  # Calculate the running time
            return [initial_state] + path, len(path), len(explored), search_depth, execution_time  # Return path, cost, nodes expanded, search depth, and running time

        explored.add(tuple(current_state))  # Add the current state to the set of explored states

        empty_tile_index = current_state.index(0)  # Find the index of the empty tile (0)
        x, y = empty_tile_index // 3, empty_tile_index % 3  # Calculate its coordinates in the 3x3 grid

        # Define possible moves: Up, Down, Left, Right
        moves = [(x - 1, y), (x + 1, y), (x, y - 1), (x, y + 1)]

        for move_x, move_y in moves:
            if 0 <= move_x < 3 and 0 <= move_y < 3:  # Check if the move is within the grid
                new_state = current_state[:]  # Create a copy of the current state
                new_empty_tile_index = move_x * 3 + move_y  # Calculate the index of the new empty tile
                new_state[empty_tile_index], new_state[new_empty_tile_index] = new_state[new_empty_tile_index], new_state[empty_tile_index]  # Swap the empty tile with a neighboring tile

                new_state_tuple = tuple(new_state)  # Convert the new state to a tuple

                if new_state_tuple not in explored and new_state_tuple not in stack_states:
                    stack.append((new_state, path + [new_state]))
                    stack_states.add(new_state_tuple)
                    search_depth = max(search_depth, len(path) + 1)

        # Check if the elapsed time exceeds 5 minutes
        if time.time() - start_time > 800:  # 300 seconds = 5 minutes
            return None  # Return None if the execution time exceeds 5 minutes
```

### 3. A* Algorithm

#### Define Manhattan and Euclidean Distances:
- The Manhattan distance heuristic calculates the sum of the Manhattan distances of each tile from its goal position. It counts the number of horizontal and vertical moves required to place each tile in its correct position.
- The Euclidean distance heuristic calculates the Euclidean distance of each tile from its goal position. It considers the straight-line distance between the current and goal positions of each tile.

```python
from queue import PriorityQueue
from math import sqrt

def manhattan_distance(state):
    distance = 0
    for i in range(9):
        tile = state[i]
        if tile != 0:
            goal_row, goal_col = divmod(tile, 3)
            current_row, current_col = divmod(i, 3)
            distance += abs(current_row - goal_row) + abs(current_col - goal_col)
    return distance

def euclidean_distance(state):
    distance = 0
    for i in range(9):
        tile = state[i]
        if tile != 0:
            goal_row, goal_col = divmod(tile, 3)
            current_row, current_col = divmod(i, 3)
            dx = current_row - goal_row
            dy = current_col - goal_col
            distance += sqrt(dx * dx + dy * dy)
    return distance
```

```python
def A_Star(initial_state, heuristic):
    
    # Define the goal state
    goal_state = [0, 1, 2, 3, 4, 5, 6, 7, 8]
    
    # Create a priority queue for A* with (priority, state, path)
    priority_queue = PriorityQueue()
    priority_queue.put((0, initial_state, []))
    
    explored = set()  # to keep track of explored states
    start_time = time.time()
    
    search_depth = 0  # Initialize the search depth
    
    while not priority_queue.empty():
        priority, current_state, path = priority_queue.get()
        
        if current_state == goal_state:
            end_time = time.time()
            execution_time = end_time - start_time
            return [initial_state] + path, len(path), len(explored), len(path), execution_time

        explored.add(tuple(current_state))
        
        empty_tile_index = current_state.index(0)  # Find the index of the empty tile (0)
        x, y = empty_tile_index // 3, empty_tile_index % 3  # Calculate its coordinates in the 3x3 grid

        # Define possible moves: Up, Down, Left, Right
        moves = [(x - 1, y), (x + 1, y), (x, y - 1), (x, y + 1)]

        for move_x, move_y in moves:
            if 0 <= move_x < 3 and 0 <= move_y < 3:  # Check if the move is within the grid
                new_state = current_state[:]  # Create a copy of the current state
                new_empty_tile_index = move_x * 3 + move_y  # Calculate the index of the new empty tile
                new_state[empty_tile_index], new_state[new_empty_tile_index] = new_state[new_empty_tile_index], new_state[empty_tile_index]  # Swap the empty tile with a neighboring tile
                
                if tuple(new_state) not in explored and not any(item[1] == new_state for item in priority_queue.queue):
                    if heuristic == 'manhattan':
                        h = manhattan_distance(new_state)
                    elif heuristic == 'euclidean':
                        h = euclidean_distance(new_state)
                    else:
                        raise ValueError("Invalid heuristic choice")
                    
                    g = len(path) + 1  # Because one new node is added
                    f = g + h  # sum of actual cost and heuristic
                    priority_queue.put((f, new_state, path + [new_state]))
                    search_depth = max(search_depth, len(path))
        # Check if the elapsed time exceeds 5 minutes
        if time.time() - start_time > 300:  # 300 seconds = 5 minutes
            return None  # Return None if the execution time exceeds 5 minutes
```

#### Draw Game Board

```python
def draw_board(state):
    board_canvas.delete("all")
    cell_width = 180  # Increase cell width
    cell_height = 180  # Increase cell height
    if state:  # Check if state is not empty
        for i in range(3):
            for j in range(3):
                cell_value = state[i * 3 + j]  # Adjust to 1D list indexing
                x0 = j * cell_width
                y0 = i * cell_height
                x1 = (j + 1) * cell_width
                y1 = (i + 1) * cell_height
                if cell_value != 0:
                    # Fill the cell with blue color
                    board_canvas.create_rectangle(x0, y0, x1, y1, outline="black", fill="blue")
                    board_canvas.create_text((x0 + x1) / 2, (y0 + y1) / 2, text=str(cell_value), font=("Helvetica", 36))
                else:
                    # Leave the blank cell empty
                    board_canvas.create_rectangle(x0, y0, x1, y1, outline="black")
```

```python
def delayed_function():
    global cost, nodes_expanded, search_depth, running_time, path, solving_message
    cost = 0
    nodes_expanded = 0
    search_depth = 0
    running_time = 0.0
    path = []  # Reset the path

    cost_var.set(f"Cost: {cost}")
    nodes_expanded_var.set(f"Nodes Expanded: {nodes_expanded}")
    search_depth_var.set(f"Search Depth: {search_depth}")
    running_time_var.set(f"Running Time: {running_time:.2f} seconds")
    solving_message = "Solving..."
    solving_message_label.config(text=solving_message)

    initial_state_str = initial_state_entry.get()
    initial_state_values = [int(val) for val in initial_state_str.split()]

    if len(initial_state_values) != 9 or initial_state_values.count(0) != 1 or not all(0 <= val <= 8 for val in initial_state_values) or len(set(initial_state_values)) != len(initial_state_values):
        solving_message_label.config(text="Invalid Input")
        messagebox.showerror("Error", "Please enter 9 distinct values for the initial state with exactly one empty tile, each value between 0 and 8 (inclusive)!")
        return

    draw_board(initial_state_values)  # Display initial state using 1D list

    # Check if the Puzzle is Solvable
    if not is_solvable(initial_state_values):  # Check solvability of 1D list
        solving_message_label.config(text="Cannot Be Solved :( ")
        messagebox.showerror("Puzzle Unsolvable", "Can't Reach The Goal State By Sliding Tiles Using Empty Space")
        return

    root.after(10, solve_puzzle)

# Function to solve the puzzle and update the board
def solve_puzzle():
    global cost, nodes_expanded, search_depth, running_time, path, solving_message

    initial_state_str = initial_state_entry.get()
    initial_state_values = [int(val) for val in initial_state_str.split()]

    selected_algorithm = algorithm_var.get()
    selected_heuristic = heuristic_var.get()

    start_time = time.time()
    if selected_algorithm == "DFS":
        result = dfs(initial_state_values)
    elif selected_algorithm == "BFS":
        result = bfs(initial_state_values)
    elif selected_algorithm == "A*":
        if selected_heuristic == "Manhattan":
            result = A_Star(initial_state_values, heuristic='manhattan')
        elif selected_heuristic == "Euclidean":
            result = A_Star(initial_state_values, heuristic='euclidean')
        else:
            messagebox.showerror("Error", "Invalid heuristic selected.")
            solving_message_label.config(text="Invalid Heuristic")
            return
    if result == None:
        solving_message_label.config(text="Time Exceeds Limit (5 minutes)")
        messagebox.showerror("Error", "Time Exceeds Limit (5 minutes)")
        return 
    if is_solvable(initial_state_values) and result != None:
        path, cost, nodes_expanded, search_depth, running_time = result
        cost_var.set(f"Cost: {cost}")
        nodes_expanded_var.set(f"Nodes Expanded: {nodes_expanded}")
        search_depth_var.set(f"Search Depth: {search_depth}")
        running_time_var.set(f"Running Time: {running_time:.2f} seconds")
        solving_message_label.config(text="Puzzle Solved!")
        global current_step
        current_step = 0
        draw_board(initial_state_values)
```
